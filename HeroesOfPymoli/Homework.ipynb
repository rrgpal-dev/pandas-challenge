```python
#Age Demographics
bins = [0,9.99,14.99,19.99,24.99,29.99,34.99,39.99,44.99,49.99]
group_names = ["<10","10-14", "15-19", "20-24", "25-29","30-34","35-39","40-44","45-49"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],bins,labels=group_names,include_lowest=True)

#Purchase Count
re_data = purchase_data.groupby("Age Group")["Purchase ID"].count()

#Average Purchase Price
re_data_avgPrice = purchase_data.groupby("Age Group")["Price"].mean()

#Total Purchase Value
re_data_total = purchase_data.groupby("Age Group")["Price"].sum()
mer = pd.merge(re_data,re_data_avgPrice,on=["Age Group"])
mer2 = pd.merge(mer,re_data_total,on=["Age Group"])
mer2



#Average Purchase Total per Person by Age Group
re_data_total_sn = purchase_data.groupby(["Age Group","SN"])["Price"].mean().dropna()
re_data_total_sn
```

```python
#Top Spenders

#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#Purchase Count
df_topa = pud.loc[:,["SN","Purchase ID"]].groupby(["SN"]).count()
  
#Average Purchase Price
df_topb = pud.loc[:,["SN","Price"]].groupby(["SN"]).mean()
#Total Purchase Value
df_topc = pud.loc[:,["SN","Price"]].groupby(["SN"]).sum()
df_top5 = pd.merge(df_topa,df_topb,on="SN")
df_top5 = pd.merge(df_top5,df_topc,on="SN")
df_top5.columns = [ "Purchase Count","Average Purchase Price","Total Purchase Value"]
df_sort_top5 = df_top5.sort_values(by=["Purchase Count","Total Purchase Value"],ascending=False).head(5)
$df_sort_top5.sort_values(by="Total Purchase Value",ascending=False)
```
