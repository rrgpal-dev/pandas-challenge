```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"
pd.options.display.float_format = '${:,.2f}'.format

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
pud =  pd.read_csv(file_to_load)
purchase_data.head()
```

```python
# Player Count
print(f"Total Number of Players:{purchase_data['SN'].nunique()}")
```

```python
#Purchasing Analysis (Total)
pur = {"Number of Unique Items":[purchase_data["Item ID"].nunique()], 
          "Average Price":[purchase_data["Price"].mean()],
          "Number of Purchases":[purchase_data["Purchase ID"].count()],
          "Total Revenue":[purchase_data["Price"].sum()]}
pur_df = pd.DataFrame(data=pur)

# Formatting of Numbers
pur_df["Average Price"] = pur_df["Average Price"].map("${0:,.2f}".format)
pur_df["Total Revenue"] = pur_df["Total Revenue"].map("${0:,.2f}".format)


pur_df.apply
print(pur_df)
```

```python
#Purchasing Analysis (Gender)
df_gen_a = pud.loc[:,["Gender","SN"]].groupby(["Gender"]).count()
#Avg Price
df_gen_b = pud.loc[:,["Gender","Price"]].groupby(["Gender"]).mean()
#Total Purchase Value
df_gen_c = pud.loc[:,["Gender","Price"]].groupby(["Gender"]).sum()
#Avg Total Purchase Value
df_gen_d = pud.loc[:,["Gender","Price"]].groupby(["Gender"]).mean()
df_gen = pd.merge(df_gen_a,df_gen_b,on=["Gender"])
df_gen = pd.merge(df_gen,df_gen_c,on=["Gender"])
df_gen = pd.merge(df_gen,df_gen_d,on=["Gender"])
df_gen.columns = ["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]
df_gen
```

```python
#Age Demographics
bins = [0,9.99,14.99,19.99,24.99,29.99,34.99,39.99,99.99]
group_names = ["<10","10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]
age_data = purchase_data[["Gender","SN","Age"]].drop_duplicates()
age_data["Age Group"] = pd.cut(age_data["Age"],bins,labels=group_names,include_lowest=True)

#Purchase Count
re_data = age_data.groupby("Age Group").count()
val = pd.to_numeric(age_data.count())
#Percent of players 
re_data["Percentage of Players"]= (re_data["Age"]/576)*100
re_data["Percentage of Players"]  = re_data["Percentage of Players"].map("{0:,.2f}%".format)

re_data.columns = ["Total Count","SN Count","Age Count","Percentage of Players"]
#Percent of players and display
re_data[["Total Count","Percentage of Players"]]

```

```python
#Top Spenders

#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#Purchase Count
df_topa = pud.loc[:,["SN","Purchase ID"]].groupby(["SN"]).count()
  
#Average Purchase Price
df_topb = pud.loc[:,["SN","Price"]].groupby(["SN"]).mean()
#Total Purchase Value
df_topc = pud.loc[:,["SN","Price"]].groupby(["SN"]).sum()
df_top5 = pd.merge(df_topa,df_topb,on="SN")
df_top5 = pd.merge(df_top5,df_topc,on="SN")
df_top5.columns = [ "Purchase Count","Average Purchase Price","Total Purchase Value"]
df_sort_top5 = df_top5.sort_values(by=["Purchase Count","Total Purchase Value"],ascending=False).head(5)
df_sort_top5 = df_sort_top5.sort_values(by="Total Purchase Value",ascending=False)
df_sort_top5  
```

```python
#Most Popular Items

#Identify the 5 most popular items by purchase count, then list (in a table):
    #Item ID
    #Item Name
    #Purchase Count
df_item_a = pud.loc[:,["Item ID","Item Name","SN"]].groupby(["Item ID","Item Name"]).count()
#Item Price
df_item_b = pud.loc[:,["Item ID","Item Name","Price"]].groupby(["Item ID","Item Name"]).mean()
#Total Purchase Value

df_item_c = pud.loc[:,["Item ID","Item Name","Price"]].groupby(["Item ID","Item Name"]).sum()

df_item = pd.merge(df_item_a,df_item_b,on=["Item ID","Item Name"])
df_item = pd.merge(df_item,df_item_c,on=["Item ID","Item Name"])
df_item.columns = ["Purchase Count","Item Price","Total Purchase Value"]

# Sort & display Most Popular Items
df_item = df_item.sort_values(by="Purchase Count",ascending=False)
df_item.head(5)
```

```python
#Most Profitable Items

#Identify the 5 most profitable items by total purchase value, then list (in a table):
# Sort & display Most Popular Items
df_item = df_item.sort_values(by="Total Purchase Value",ascending=False)
df_item.head(5)
```
